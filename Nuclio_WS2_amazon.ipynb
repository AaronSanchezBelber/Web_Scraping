```python
keyword="iphone"
```

```python
driver.get("https://www.amazon.es/s?k="+keyword)
```

```python
content =driver.page_source
```

```python
soup = BeautifulSoup(content)
```

```python
soup
```

```python
resultado = soup.findAll(class_ = "a-section a-spacing-base")
```

```python
resultado
```

```python
if len(resultado) ==0:
    resultado = soup.findAll(class_ ="a-section a-spacing-medium a-text-center")
```

```python
df_lista_productos = pd.DataFrame(columns=["titulo","precio","imagen"])
```

```python
if not os.path.exists("/content/drive/MyDrive/Colab Notebooks/Clase 3 - Web scraping/WS_"+keyword):
    os.makedirs("/content/drive/MyDrive/Colab Notebooks/Clase 3 - Web scraping/WS_" +keyword)
```

```python
carpeta = "/content/drive/MyDrive/Colab Notebooks/Clase 3 - Web scraping/WS_"+keyword+"/"
```

```python
for elemento in resultado:
    titulo_producto = elemento.find(class_= "a-size-base-plus a-color-base a-text-normal").get_text()
    precio_producto= elemento.find(class_= "a-price-whole")
    imagen_producto = elemento.find("img")
    if precio_producto != None:
        precio_producto= precio_producto.get_text()
    else:
        precio_producto= "No existe precio"

    fichero = open(carpeta+basename(imagen_producto.attrs["src"]),"wb")
    fichero.write(requests.get(imagen_producto.attrs["src"]).content)

    df_lista_productos = df_lista_productos.append({"titulo":titulo_producto, "precio":precio_producto, "imagen":"<img src='"+imagen_producto.attrs["src"]+"'>"},
    ignore_index=True)
```

```python
df_lista_productos.to_csv(carpeta+keyword+".csv")
```

```python
df_lista_productos.to_html(carpeta+keyword+".html", escape=False)
```
